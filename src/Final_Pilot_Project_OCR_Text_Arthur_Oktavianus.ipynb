```python
# Import library
import fitz  # PyMuPDF untuk membaca PDF
import pytesseract  # OCR untuk membaca teks dari gambar
from PIL import Image  # Untuk memproses gambar
import re
import numpy as np
import networkx as nx
import nltk
import openai
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from scipy.spatial.distance import cosine
from transformers import pipeline  # Model NLP untuk tanya jawab


# Unduh resource yang diperlukan untuk NLTK
nltk.download('punkt')
nltk.download('stopwords')
stopwords = set(stopwords.words("indonesian"))

# Load model untuk tanya jawab (Q&A)
qa_pipeline = pipeline("question-answering", model="bert-large-uncased-whole-word-masking-finetuned-squad")

# Fungsi untuk membaca teks dari PDF
def read_pdf(file_path):
    """Membaca teks dari file PDF"""
    doc = fitz.open(file_path)
    text = ""
    for page in doc:
        text += page.get_text("text") + "\n"
    return text

# Fungsi untuk membaca teks dari gambar (OCR)
def read_image(image_path):
    """Membaca teks dari gambar menggunakan OCR"""
    img = Image.open(image_path)
    text = pytesseract.image_to_string(img, lang="ind")  # Menggunakan bahasa Indonesia
    return text

# Fungsi untuk membersihkan dan tokenisasi teks
def read_article(file_text):
    """Membersihkan teks dan melakukan tokenisasi kalimat"""
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols  & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
    file_text = emoji_pattern.sub(r'', file_text)
    tokens = nltk.tokenize.sent_tokenize(file_text)
    sentences = [sentence.replace("[^a-zA-Z]", " ").split(" ") for sentence in tokens]
    return sentences

# Fungsi untuk menghitung kemiripan kalimat
def sentence_similarity(sent1, sent2, stopwords=None):
    if stopwords is None:
        stopwords = []

    sent1 = [w.lower() for w in sent1]
    sent2 = [w.lower() for w in sent2]
    all_words = list(set(sent1 + sent2))

    vector1 = [0] * len(all_words)
    vector2 = [0] * len(all_words)

    for w in sent1:
        if w in stopwords:
            continue
        vector1[all_words.index(w)] += 1

    for w in sent2:
        if w in stopwords:
            continue
        vector2[all_words.index(w)] += 1

    return 1 - cosine(vector1, vector2)

# Fungsi untuk menghasilkan matriks kesamaan antar kalimat
def gen_sim_matrix(sentences, stop_words):
    similarity_matrix = np.zeros((len(sentences), len(sentences)))
    for idx1 in range(len(sentences)):
        for idx2 in range(len(sentences)):
            if idx1 == idx2:
                continue
            similarity_matrix[idx1][idx2] = sentence_similarity(sentences[idx1], sentences[idx2], stop_words)
    return similarity_matrix

# Fungsi utama untuk membaca file dan membuat ringkasan
def generate_summary(file_path, top_n):
    """Membaca file (PDF atau Gambar), mengekstrak teks, dan membuat ringkasan"""
    # Tentukan apakah file adalah PDF atau Gambar
    if file_path.lower().endswith(".pdf"):
        file_text = read_pdf(file_path)
    elif file_path.lower().endswith((".png", ".jpg", ".jpeg")):
        file_text = read_image(file_path)
    else:
        raise ValueError("Format file tidak didukung. Harap gunakan PDF atau Gambar (PNG, JPG, JPEG).")

    # Proses teks menjadi ringkasan
    sentences = read_article(file_text)

    # Batasi top_n agar tidak melebihi 100%
    top_n = min(top_n, 100)

    persentase = top_n / 100 * len(sentences)

    stop_words = stopwords  # Gunakan set stopwords yang sudah didefinisikan sebelumnya
    summarize_text = []
    sentence_similarity_matrix = gen_sim_matrix(sentences, stop_words)
    sentence_similarity_graph = nx.from_numpy_array(sentence_similarity_matrix)
    scores = nx.pagerank(sentence_similarity_graph)
    ranked_sentence = sorted(
        ((scores[i], s) for i, s in enumerate(sentences)), reverse=True
    )

    # fungsi ini untuk memastikan tidak melebihi jumlah kalimat yang tersedia
    for i in range(min(round(persentase), len(ranked_sentence))):
        summarize_text.append(" ".join(ranked_sentence[i][1]))

    array_json = ' '.join(summarize_text)
    data = {'summary': array_json, 'full_text': file_text}
    return data

# Fungsi untuk tanya jawab berdasarkan teks yang sudah diekstrak
def ask_question(text, question):
    """Menjawab pertanyaan berdasarkan teks yang diberikan"""
    result = qa_pipeline(question=question, context=text)
    return result["answer"]
```

```python
# ======= PANGGIL FUNGSI UNTUK MEMBUAT RINGKASAN & TANYA JAWAB =======

# Gunakan salah satu file ini (ganti dengan path yang sesuai)
file_path = "/content/Tim Indonesia_png.png"  # Untuk PDF

# Jumlah persentase kalimat yang ingin diringkas (misalnya 50%)
top_n = 50

# Jalankan ringkasan
summary_result = generate_summary(file_path, top_n)

# Tampilkan hasil ringkasan
print("=== HASIL RINGKASAN ===")
print(summary_result['summary'])
```

```python
# Tanya jawab
print("\n=== TANYA JAWAB ===")
question = "Berapa skor pertandingan tim Hindia Belanda melawan tim nasional Singapura?"
answer = ask_question(summary_result['full_text'], question)
print(f"Pertanyaan: {question}")
print(f"Jawaban: {answer}")
```

```python
# Masukkan API Key OpenAI Anda di sini
OPENAI_API_KEY = "XXXXXX"

# Inisialisasi OpenAI client
client = openai.OpenAI(api_key=OPENAI_API_KEY)

def ask_chatgpt(context, question):
    """Menggunakan GPT-4o untuk menjawab pertanyaan berdasarkan teks yang diekstrak"""
    prompt = f"Berikut adalah teks:\n\n{context}\n\nJawablah pertanyaan berikut dengan jelas:\n{question}"

    response = client.chat.completions.create(
        model="gpt-4o",  # Gunakan GPT-4o untuk jawaban lebih baik
        messages=[{"role": "system", "content": "Anda adalah asisten AI yang membantu menjawab pertanyaan berdasarkan teks yang diberikan."},
                  {"role": "user", "content": prompt}],
        temperature=0.7,
        max_tokens=300
    )

    return response.choices[0].message.content
```

```python
print("\n=== TANYA JAWAB MENGGUNAKAN GPT-4o ===")
question = "Berapa skor pertandingan antara hindia belanda dengan tim nasional singapura?"
answer_gpt = ask_chatgpt(summary_result['full_text'], question)
print(f"Pertanyaan: {question}")
print(f"Jawaban (GPT-4o): {answer_gpt}")
```
